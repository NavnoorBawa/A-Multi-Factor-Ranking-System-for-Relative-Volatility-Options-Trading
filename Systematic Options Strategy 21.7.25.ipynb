```python
#!/usr/bin/env python3
"""
FORCED ACTIVITY Rough Volatility Options Trading System
========================================================

This version dramatically relaxes all filters to guarantee trade execution,
addressing the issue of zero trades in previous versions. The goal is to
force activity to generate a performance baseline for future refinement.

Version 18.0 - FORCED ACTIVITY:
- EXTREMELY RELAXED THRESHOLDS: Volatility, liquidity, and price filters are
  set to minimal levels to ensure signals are generated.
- NO SPREAD OR LIQUIDITY GATES: The bid-ask spread and initial liquidity
  filters have been removed to maximize the pool of potential options.
- RANK-AND-SELECT REMAINS: The system still scores and ranks all qualifying
  opportunities to select the best trade available.

Author: Advanced Quantitative Trading Research
Version: 18.0 - Forced Activity Edition
Date: 2025-07-21
"""

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import cm
import seaborn as sns
from scipy import optimize, stats
from scipy.stats import norm
from scipy.linalg import cholesky
from scipy.special import gamma as gamma_func
import yfinance as yf
from datetime import datetime, timedelta
import logging
import json
import time
import sys
from typing import Dict, List, Tuple, Optional, Union
from dataclasses import dataclass, asdict
import traceback

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")

@dataclass
class BacktestMetrics:
    """Enhanced metrics for the event-driven backtest performance"""
    start_date: str
    end_date: str
    total_return_pct: float
    annualized_return_pct: float
    annualized_volatility_pct: float
    sharpe_ratio: float
    max_drawdown_pct: float
    calmar_ratio: float
    sortino_ratio: float
    win_rate_pct: float
    avg_trade_pnl: float
    best_trade: float
    worst_trade: float
    num_trades: int
    total_costs: float
    pct_days_active: float
    final_equity: float
    profit_factor: float

class BlackScholes:
    """Enhanced Black-Scholes pricing with numerical stability"""

    @staticmethod
    def price(S, K, T, r, sigma, option_type='call'):
        """Calculate Black-Scholes option price with enhanced stability"""
        if np.any(T <= 1e-6):
            if option_type == 'call':
                return np.maximum(0, S - K)
            else:
                return np.maximum(0, K - S)
        sigma = np.maximum(sigma, 1e-4)
        with np.errstate(divide='ignore', invalid='ignore'):
            d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
            d2 = d1 - sigma * np.sqrt(T)
        if option_type == 'call':
            price = (S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2))
        else:
            price = (K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1))
        intrinsic = np.maximum(0, S - K) if option_type == 'call' else np.maximum(0, K - S)
        return np.where(np.isnan(price) | (price < intrinsic), intrinsic, price)

    @staticmethod
    def delta(S, K, T, r, sigma, option_type='call'):
        """Calculate option delta"""
        if T <= 1e-6:
            if option_type == 'call':
                return 1.0 if S > K else 0.0
            else:
                return -1.0 if S < K else 0.0
        sigma = np.maximum(sigma, 1e-4)
        with np.errstate(divide='ignore', invalid='ignore'):
            d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
        if option_type == 'call':
            return norm.cdf(d1)
        else:
            return norm.cdf(d1) - 1

class AggressiveTradingSystem:
    """
    FORCED-ACTIVITY trading system using a rank-based selection model.
    """

    def __init__(self):
        self.recent_vol_window = 20

    def calculate_historical_volatility(self, prices: pd.Series, window: int = 20) -> float:
        """Calculate annualized historical volatility"""
        if len(prices) < window: return 0.20
        returns = np.log(prices / prices.shift(1)).dropna()
        if len(returns) < 2: return 0.20
        vol = returns.tail(window).std() * np.sqrt(252)
        return np.clip(vol, 0.05, 1.5)

    def generate_signals(self, market_data: Dict, historical_prices: pd.Series) -> List[Dict]:
        """
        Generates signals using extremely relaxed filters to force activity.
        """
        signals = []
        if not market_data or market_data.get('calls') is None or market_data.get('puts') is None:
            return signals

        try:
            current_price = market_data['current_price']
            hist_vol = self.calculate_historical_volatility(historical_prices)
            vol_rank = market_data.get('vol_rank', 0.5)
            sma_20 = market_data.get('sma_20', current_price)
            
            is_uptrend = current_price > sma_20
            is_high_iv_rank = vol_rank > 0.5
            is_low_iv_rank = vol_rank < 0.5

            # EXTREMELY RELAXED thresholds to force trades
            iv_sell_threshold = 1.01 # Tiny 1% edge
            iv_buy_threshold = 0.99  # Tiny 1% edge

            for option_type in ['calls', 'puts']:
                options_df = market_data[option_type].copy()
                if options_df.empty: continue

                # REMOVED most filters to guarantee signals are found
                options_df = options_df[
                    (options_df['mid_price'] > 0.01) & # Almost no price filter
                    (abs(options_df['strike'] / current_price - 1) < 0.25) # Very wide moneyness
                ].copy()

                if options_df.empty: continue

                for idx, option in options_df.iterrows():
                    market_iv = option['impliedVolatility']
                    iv_ratio = market_iv / hist_vol if hist_vol > 0.01 else 1.0

                    base_confidence = 0.0
                    action = None

                    # Step 1: Identify a core volatility signal
                    if iv_ratio > iv_sell_threshold:
                        action = 'sell'
                        base_confidence = (iv_ratio - iv_sell_threshold) * 5.0 # Higher multiplier
                    elif iv_ratio < iv_buy_threshold:
                        action = 'buy'
                        base_confidence = (iv_buy_threshold - iv_ratio) * 5.0 # Higher multiplier
                    
                    if not action: continue

                    # Step 2: Score the signal based on confirming factors
                    trend_bonus = 0.0
                    vol_regime_bonus = 0.0

                    if action == 'sell':
                        if is_high_iv_rank: vol_regime_bonus = 0.25 # Higher bonus
                        if (is_uptrend and option_type == 'puts') or \
                           (not is_uptrend and option_type == 'calls'):
                            trend_bonus = 0.25 # Higher bonus
                    elif action == 'buy':
                        if is_low_iv_rank: vol_regime_bonus = 0.25 # Higher bonus
                        if (is_uptrend and option_type == 'calls') or \
                           (not is_uptrend and option_type == 'puts'):
                            trend_bonus = 0.25 # Higher bonus
                    
                    final_confidence = base_confidence + trend_bonus + vol_regime_bonus
                    
                    # Step 3: Add any valid signal to the list for ranking
                    entry_price = option['bid'] if action == 'sell' else option['ask']
                    if entry_price > 0:
                        signals.append({
                            'date': market_data['date'], 'strike': option['strike'],
                            'option_type': option_type[:-1], 'action': action,
                            'entry_price': entry_price, 'confidence': final_confidence,
                            'iv_ratio': iv_ratio, 'hist_vol': hist_vol,
                            'market_iv': market_iv, 'moneyness': option['strike'] / current_price
                        })

            # Step 4: Rank all generated signals and return the best ones
            signals.sort(key=lambda x: x['confidence'], reverse=True)
            return signals[:3] # Return the top 3 signals for consideration

        except Exception as e:
            logger.error(f"Signal generation error: {e}\n{traceback.format_exc()}")
            return []

class MarketDataSimulator:
    """Realistic market data simulator with proper option chains"""
    def __init__(self, symbol: str, start_date: str, end_date: str):
        logger.info(f"Initializing MarketDataSimulator for {symbol}...")
        self.hist_data = self._fetch_hist_data(symbol, start_date, end_date)
        self._precompute_indicators()
        logger.info("Market data simulation ready.")

    def _fetch_hist_data(self, symbol, start_date, end_date):
        data = yf.Ticker(symbol).history(start=pd.to_datetime(start_date) - timedelta(days=365), end=end_date)
        if data.empty: raise ValueError(f"Could not fetch historical price data for {symbol}")
        data.index = data.index.tz_localize(None)
        return data

    def _precompute_indicators(self):
        self.hist_data['log_return'] = np.log(self.hist_data['Close'] / self.hist_data['Close'].shift(1))
        self.hist_data['realized_vol'] = self.hist_data['log_return'].rolling(window=20).std() * np.sqrt(252)
        self.hist_data['sma_20'] = self.hist_data['Close'].rolling(window=20).mean()
        self.hist_data['vol_rank'] = self.hist_data['realized_vol'].rolling(window=252).rank(pct=True)

    def get_market_for_day(self, date: pd.Timestamp) -> Optional[Dict]:
        if date not in self.hist_data.index: return None
        row = self.hist_data.loc[date]
        current_price = row['Close']
        strike_step = max(1, round(current_price / 100))
        range_pct = 0.25 # Wider range for more options
        strike_min = int(current_price * (1 - range_pct) / strike_step) * strike_step
        strike_max = int(current_price * (1 + range_pct) / strike_step) * strike_step
        strikes = np.arange(strike_min, strike_max + strike_step, strike_step)
        atm_vol = row.get('realized_vol', 0.20)
        if pd.isna(atm_vol) or atm_vol < 0.05: atm_vol = 0.20
        moneyness = strikes / current_price
        smile = atm_vol + 0.10 * (moneyness - 1)**2
        vols = np.clip(smile, 0.08, 1.50)
        T = 30 / 365.0
        r = 0.05
        mid_prices_call = BlackScholes.price(current_price, strikes, T, r, vols, 'call')
        mid_prices_put = BlackScholes.price(current_price, strikes, T, r, vols, 'put')
        spread_pct = 0.015 # Realistic but not overly punitive spread
        distance = np.abs(strikes - current_price) / current_price
        volume_base = np.exp(-15 * distance) * 150
        volume = np.random.poisson(volume_base) + 1 # Ensure at least 1
        calls_df = pd.DataFrame({'strike': strikes, 'mid_price': mid_prices_call, 'bid': mid_prices_call * (1 - spread_pct), 'ask': mid_prices_call * (1 + spread_pct), 'impliedVolatility': vols, 'volume': volume, 'openInterest': volume * np.random.uniform(2, 5, len(strikes)), 'optionType': 'call'})
        puts_df = pd.DataFrame({'strike': strikes, 'mid_price': mid_prices_put, 'bid': mid_prices_put * (1 - spread_pct), 'ask': mid_prices_put * (1 + spread_pct), 'impliedVolatility': vols, 'volume': volume, 'openInterest': volume * np.random.uniform(2, 5, len(strikes)), 'optionType': 'put'})
        return {'date': date, 'current_price': current_price, 'time_to_expiration': T, 'calls': calls_df, 'puts': puts_df, 'vol_rank': row.get('vol_rank', 0.5), 'sma_20': row.get('sma_20', current_price)}

class PortfolioManager:
    """
    Portfolio Manager with intelligent risk management (stops, targets, holding period).
    """
    def __init__(self, initial_equity: float):
        self.equity = initial_equity
        self.initial_equity = initial_equity
        self.positions = []
        self.trade_log = []
        
        self.max_open_positions = 10
        self.risk_per_trade_pct = 0.02
        self.commission_per_contract = 0.65
        
        # Smart Exit Rules
        self.holding_period = 5
        self.stop_loss_multiplier_short = 2.0 # Stop loss for short options (premium doubles)
        self.stop_loss_pct_long = 0.50 # 50% stop loss for long options
        self.profit_target_multiplier_short = 0.5 # 50% profit target for short options
        self.profit_target_pct_long = 1.0 # 100% profit target for long options (doubles)

    def close_positions(self, date: pd.Timestamp, market_data: Dict):
        """Close positions based on smart exit rules."""
        if not self.positions or not market_data: return

        closed_positions = []
        for pos in self.positions[:]:
            try:
                days_held = (date - pos['date']).days
                options_df = market_data.get(f"{pos['option_type']}s", pd.DataFrame())
                if options_df.empty: continue
                
                matching_option = options_df[options_df['strike'] == pos['strike']]
                if matching_option.empty: continue
                
                option_data = matching_option.iloc[0]
                exit_price = option_data['ask'] if pos['action'] == 'sell' else option_data['bid']
                if exit_price <= 0: continue

                reason_for_exit = None
                if pos['action'] == 'sell':
                    if exit_price >= pos['entry_price'] * self.stop_loss_multiplier_short: reason_for_exit = "Stop-Loss"
                    elif exit_price <= pos['entry_price'] * self.profit_target_multiplier_short: reason_for_exit = "Profit-Target"
                else: # Buy
                    if exit_price <= pos['entry_price'] * (1 - self.stop_loss_pct_long): reason_for_exit = "Stop-Loss"
                    elif exit_price >= pos['entry_price'] * (1 + self.profit_target_pct_long): reason_for_exit = "Profit-Target"
                
                if days_held >= self.holding_period and not reason_for_exit:
                    reason_for_exit = "Max Hold"
                
                if not reason_for_exit: continue

                pnl_per_contract = (pos['entry_price'] - exit_price) if pos['action'] == 'sell' else (exit_price - pos['entry_price'])
                pnl = pnl_per_contract * pos['size'] * 100
                total_commission = 2 * self.commission_per_contract * pos['size']
                pnl -= total_commission
                self.equity += pnl
                
                trade_record = pos.copy()
                trade_record.update({'exit_date': date, 'exit_price': exit_price, 'pnl': pnl, 'commission': total_commission, 'days_held': days_held, 'exit_reason': reason_for_exit})
                self.trade_log.append(trade_record)
                closed_positions.append(pos)
                
                logger.info(f"Closed {pos['action']} {pos['size']} {pos['option_type']} {pos['strike']} for ${pnl:.2f} P&L ({reason_for_exit})")
            except Exception as e:
                logger.error(f"Error closing position {pos}: {e}")
                continue
        
        self.positions = [p for p in self.positions if p not in closed_positions]

    def enter_new_positions(self, date: pd.Timestamp, signals: List[Dict]):
        """Enter new positions based on high-quality signals."""
        if not signals: return
        
        for signal in signals:
            if len(self.positions) >= self.max_open_positions: break
            is_duplicate = any(p['strike'] == signal['strike'] and p['option_type'] == signal['option_type'] for p in self.positions)
            if is_duplicate: continue

            risk_amount = self.equity * self.risk_per_trade_pct
            option_premium = signal['entry_price']
            if option_premium <= 0: continue
            
            risk_per_contract = 0
            if signal['action'] == 'sell':
                risk_per_contract = (option_premium * self.stop_loss_multiplier_short - option_premium) * 100
            else: # Buy
                risk_per_contract = (option_premium * self.stop_loss_pct_long) * 100
            
            if risk_per_contract <= 0: continue
            size = max(1, int(risk_amount / risk_per_contract))
            
            max_size = 3 if signal['action'] == 'sell' else 5
            size = min(size, max_size)
            
            position = signal.copy()
            position['size'] = size
            self.positions.append(position)
            logger.info(f"Entered {signal['action']} {size} {signal['option_type']} {signal['strike']} @ ${signal['entry_price']:.2f} (conf: {signal['confidence']:.2f})")

class EventDrivenBacktester:
    """
    Backtester for the forced-activity strategy.
    """
    def __init__(self, symbol: str, start_date: str, end_date: str, initial_equity: float = 100000.0):
        self.symbol = symbol
        self.start_date = pd.to_datetime(start_date)
        self.end_date = pd.to_datetime(end_date)
        self.initial_equity = initial_equity
        self.trading_system = AggressiveTradingSystem()
        self.market_sim = MarketDataSimulator(symbol, start_date, end_date)
        self.portfolio_manager = PortfolioManager(initial_equity)
        self.equity_curve = pd.DataFrame(columns=['date', 'equity'])
        self.days_traded = 0

    def run_backtest(self):
        """Runs the event-driven backtest."""
        self.equity_curve.loc[0] = [self.start_date - timedelta(days=1), self.initial_equity]
        trading_days = self.market_sim.hist_data.loc[self.start_date:self.end_date].index
        total_days = len(trading_days)
        logger.info(f"Starting FORCED ACTIVITY backtest for {total_days} trading days...")
        
        for i, current_day in enumerate(trading_days):
            progress = int(50 * (i + 1) / total_days)
            bar = '█' * progress + '-' * (50 - progress)
            sys.stdout.write(f'\rBacktest Progress: [{bar}] {i+1}/{total_days} days')
            sys.stdout.flush()
            
            market_data = self.market_sim.get_market_for_day(current_day)
            
            self.portfolio_manager.close_positions(current_day, market_data)
            
            if not market_data:
                self.equity_curve.loc[len(self.equity_curve)] = [current_day, self.portfolio_manager.equity]
                continue
            
            self.days_traded += 1
            try:
                historical_prices = self.market_sim.hist_data['Close'].loc[:current_day].tail(50)
                signals = self.trading_system.generate_signals(market_data, historical_prices)
                self.portfolio_manager.enter_new_positions(current_day, signals)
            except Exception as e:
                logger.error(f"Trading error on {current_day.date()}: {e}\n{traceback.format_exc()}")
            
            self.equity_curve.loc[len(self.equity_curve)] = [current_day, self.portfolio_manager.equity]

        sys.stdout.write('\n')
        logger.info("FORCED ACTIVITY backtest completed.")
        return self.generate_enhanced_report()

    def generate_enhanced_report(self):
        """Generates a comprehensive performance report and plots."""
        if len(self.equity_curve) < 2:
            logger.error("Insufficient data for report generation.")
            return None
        self.equity_curve.set_index('date', inplace=True)
        self.equity_curve['returns'] = self.equity_curve['equity'].pct_change().fillna(0)
        total_return = (self.equity_curve['equity'].iloc[-1] / self.initial_equity) - 1
        num_days = (self.end_date - self.start_date).days
        annualized_return = (1 + total_return)**(365 / num_days) - 1 if num_days > 0 else 0
        annualized_vol = self.equity_curve['returns'].std() * np.sqrt(252)
        sharpe_ratio = annualized_return / annualized_vol if annualized_vol > 1e-6 else 0
        negative_returns = self.equity_curve['returns'][self.equity_curve['returns'] < 0]
        downside_deviation = negative_returns.std() * np.sqrt(252) if len(negative_returns) > 1 else 1e-6
        sortino_ratio = annualized_return / downside_deviation if downside_deviation > 0 else 0
        rolling_max = self.equity_curve['equity'].cummax()
        daily_drawdown = self.equity_curve['equity'] / rolling_max - 1.0
        max_drawdown = daily_drawdown.min()
        calmar_ratio = annualized_return / abs(max_drawdown) if max_drawdown < -1e-6 else 0
        closed_trades = pd.DataFrame(self.portfolio_manager.trade_log)
        if not closed_trades.empty:
            win_rate = (closed_trades['pnl'] > 0).mean() * 100
            avg_trade_pnl = closed_trades['pnl'].mean()
            best_trade = closed_trades['pnl'].max()
            worst_trade = closed_trades['pnl'].min()
            winning_trades = closed_trades[closed_trades['pnl'] > 0]['pnl'].sum()
            losing_trades = abs(closed_trades[closed_trades['pnl'] < 0]['pnl'].sum())
            profit_factor = winning_trades / losing_trades if losing_trades > 0 else np.inf
        else:
            win_rate = avg_trade_pnl = best_trade = worst_trade = profit_factor = 0
        metrics = BacktestMetrics(start_date=self.start_date.strftime('%Y-%m-%d'), end_date=self.end_date.strftime('%Y-%m-%d'), total_return_pct=total_return * 100, annualized_return_pct=annualized_return * 100, annualized_volatility_pct=annualized_vol * 100, sharpe_ratio=sharpe_ratio, max_drawdown_pct=max_drawdown * 100, calmar_ratio=calmar_ratio, sortino_ratio=sortino_ratio, win_rate_pct=win_rate, avg_trade_pnl=avg_trade_pnl, best_trade=best_trade, worst_trade=worst_trade, num_trades=len(closed_trades), total_costs=closed_trades['commission'].sum() if not closed_trades.empty else 0, pct_days_active=(self.days_traded / len(self.equity_curve)) * 100 if len(self.equity_curve) > 0 else 0, final_equity=self.equity_curve['equity'].iloc[-1], profit_factor=profit_factor)
        self.print_enhanced_report(metrics)
        self.plot_enhanced_results(metrics)
        return asdict(metrics)

    def print_enhanced_report(self, metrics: BacktestMetrics):
        print("\n" + "="*90)
        print("🚀 FORCED ACTIVITY TRADING SYSTEM - v18.0")
        print("="*90)
        print(f"Period: {metrics.start_date} to {metrics.end_date} | Symbol: {self.symbol}")
        print(f"Initial Equity: ${self.initial_equity:,.2f} | Final Equity: ${metrics.final_equity:,.2f}")
        print("-" * 90)
        print(f"📈 RETURNS: Total={metrics.total_return_pct:+.2f}% | Annualized={metrics.annualized_return_pct:+.2f}%")
        print(f"⚡ RISK: Volatility={metrics.annualized_volatility_pct:.2f}% | Max Drawdown={metrics.max_drawdown_pct:.2f}%")
        print(f"📊 RATIOS: Sharpe={metrics.sharpe_ratio:.3f} | Sortino={metrics.sortino_ratio:.3f} | Calmar={metrics.calmar_ratio:.3f}")
        print("-" * 90)
        print("🎯 TRADING ACTIVITY")
        print(f"Total Trades: {metrics.num_trades} | Win Rate: {metrics.win_rate_pct:.2f}% | Profit Factor: {metrics.profit_factor:.2f}")
        print(f"Avg P&L: ${metrics.avg_trade_pnl:+,.2f} | Best: ${metrics.best_trade:+,.2f} | Worst: ${metrics.worst_trade:+,.2f}")
        print(f"Strategy Active: {metrics.pct_days_active:.1f}% of days | Total Costs: ${metrics.total_costs:,.2f}")
        print("-" * 90)
        if metrics.num_trades == 0: performance_grade = "❌ NO TRADES - Model is broken or market was flat"
        elif metrics.profit_factor > 1.5 and metrics.win_rate_pct > 55: performance_grade = "🏆 EXCELLENT"
        elif metrics.profit_factor > 1.1 and metrics.win_rate_pct > 50: performance_grade = "✅ PROMISING"
        elif metrics.profit_factor > 0.9: performance_grade = "⚠️ MARGINAL - BREAKEVEN"
        else: performance_grade = "🔥 NEEDS REFINEMENT"
        print(f"📊 OVERALL PERFORMANCE: {performance_grade}")
        print("="*90 + "\n")

    def plot_enhanced_results(self, metrics: BacktestMetrics):
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 14), gridspec_kw={'height_ratios': [2, 1]})
        fig.suptitle(f'Forced Activity Strategy Performance (v18.0) - {self.symbol}', fontsize=20, fontweight='bold')
        self.equity_curve['equity'].plot(ax=ax1, linewidth=2.5, color='darkblue', label='Equity')
        ax1.axhline(y=self.initial_equity, color='gray', linestyle='--', alpha=0.7, label='Initial Equity')
        ax1.set_title(f'Portfolio Equity Curve - Final: ${metrics.final_equity:,.2f}', fontsize=14, fontweight='bold')
        ax1.set_ylabel('Portfolio Value ($)', fontsize=12)
        ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
        ax1.legend()
        rolling_max = self.equity_curve['equity'].cummax()
        drawdown = (self.equity_curve['equity'] / rolling_max - 1) * 100
        drawdown.plot(ax=ax2, color='red', linewidth=2)
        ax2.fill_between(drawdown.index, drawdown, 0, color='red', alpha=0.2)
        ax2.set_title(f'Drawdown Analysis - Max DD: {metrics.max_drawdown_pct:.2f}%', fontsize=14, fontweight='bold')
        ax2.set_ylabel('Drawdown (%)', fontsize=12)
        ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
        if self.portfolio_manager.trade_log:
            trades_df = pd.DataFrame(self.portfolio_manager.trade_log)
            sns.countplot(data=trades_df, x='exit_reason', ax=ax3, palette='viridis')
            ax3.set_title('Trade Exit Reasons', fontsize=14, fontweight='bold')
            ax3.set_xlabel('Exit Reason', fontsize=12)
            ax3.set_ylabel('Number of Trades', fontsize=12)
            sns.histplot(data=trades_df, x='pnl', ax=ax4, bins=30, kde=True, color='darkorange', edgecolor='black', line_kws={'linewidth': 2})
            ax4.axvline(trades_df['pnl'].mean(), color='red', linestyle='--', linewidth=2, label=f'Avg: ${trades_df["pnl"].mean():.2f}')
            ax4.set_title(f'Trade P&L Distribution (n={metrics.num_trades})', fontsize=14, fontweight='bold')
            ax4.set_xlabel('Trade P&L ($)', fontsize=12)
            ax4.legend()
        else:
            for ax in [ax3, ax4]:
                ax.text(0.5, 0.5, 'No Trades Executed', ha='center', va='center', transform=ax.transAxes, fontsize=16, color='red')
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

def main():
    """Main execution function for the forced-activity strategy."""
    print("🚀 FORCED ACTIVITY ROUGH VOLATILITY OPTIONS TRADING SYSTEM v18.0")
    print("=" * 100)
    print("💥 EXTREMELY RELAXED FILTERS ENABLED:")
    print("   ✅ Minimal thresholds to ensure trades are generated.")
    print("   ✅ Ranks all opportunities and selects the top 2-3 per day.")
    print("   ✅ This version prioritizes activity over selectivity.")
    print("🎯 EXPECTED ACTIVITY:")
    print("   💰 A high number of trades to create a performance baseline.")
    print("=" * 100)
    
    end_date = datetime.now()
    start_date = end_date - timedelta(days=90)
    
    backtester = EventDrivenBacktester(
        symbol='SPY', 
        start_date=start_date.strftime('%Y-%m-%d'), 
        end_date=end_date.strftime('%Y-%m-%d'),
        initial_equity=100000.0
    )
    
    try:
        results = backtester.run_backtest()
        if results:
            print(f"\n🎉 FORCED ACTIVITY BACKTEST COMPLETED!")
            if results['num_trades'] > 0:
                print(f"📊 EXECUTED {results['num_trades']} TRADES.")
                if results['profit_factor'] > 1.0:
                    print(f"💰 PROFITABLE: Profit Factor = {results['profit_factor']:.2f}")
                else:
                    print(f"⚠️ NEEDS REFINEMENT: Profit Factor = {results['profit_factor']:.2f}")
            else:
                print("🚨 NO TRADES EXECUTED - Check base thresholds or data simulation.")
        else:
            print("\n❌ Backtest failed to produce results.")
    except Exception as e:
        logger.error(f"Backtest execution failed: {e}\n{traceback.format_exc()}")

if __name__ == "__main__":
    main()
```
